```python
import csv
filename = '2022_LoL_esports_match_data_from_OraclesElixir_20221107.csv'
game_rows = []
leagues = set(['WCS', 'MSI', 'LCSA', 'LCK', 'LCK CL', 'CBLOL', 'LCS', 'LEC', 'LPL', 'Proving Grounds Circuit', 'PCS', 'VCS', 'LCL', 'LJL', 'LLA', 'TCL'])
with open(filename) as f:
    reader = csv.DictReader(f)
    for row in reader:
        if row['league'] in leagues:
            game_rows.append(row)
```

```python
champs = set()
games = {}
for record in game_rows:
    game_team = str(record['gameid']) + str(record['teamid'])
    if record['champion'] == '':
        continue
    elif game_team in games:
        games[game_team]['champs'].append(record['champion'])
    else:
        games[game_team] = {'champs':[record['champion']], 'result': record['result']}   
    champs.add(record['champion'])
```

```python
pair_counts = {}
for k, v in games.items():
    ps = list(combinations(v['champs'], 2))
    for p in ps:
        p = tuple(sorted(p))
        if p in pair_counts:
            pair_counts[p]['count'] += 1
            pair_counts[p]['result'] += int(v['result'])
        else:
            pair_counts[p] = {'count':1, 'result': int(v['result'])}
```

```python
for k, v in pair_counts.copy().items():
    if pair_counts[k]['count'] < 15:
        pair_counts.pop(k)
    else:
        pair_counts[k] = pair_counts[k]['result'] / pair_counts[k]['count']
    
```
